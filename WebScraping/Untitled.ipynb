```python
final=pd.DataFrame()

for j in range(7,10):
    url=requests.get('https://medex.com.bd/brands?page={}'.format(j)).text
    soup=BeautifulSoup(url)
    medicine=soup.find_all('div',class_='row data-row')
    name=[]
    weight=[]
    g_name=[]
    companyname=[]
    for med in medicine:
        try:
            name.append(med.find('div','col-xs-12 data-row-top').text.strip())
        except:
            name.append(np.nan)
        try:
            weight.append(med.find('div','col-xs-12 data-row-strength').text.strip())
        except:
            weight.append(np.nan)
        try:
            g_name.append(med.find_all('div','col-xs-12')[2].text.strip())
        except:
            g_name.append(np.nan)
        try:
            companyname.append(med.find('span','data-row-company').text.strip())
        except:
            companyname.append(np.nan)
    df =pd.DataFrame({
    'name':name,
    'weight':weight,
    'g_name':g_name,
    'companyname':companyname
    })
    
    final=final._append(df,ignore_index=True)
#     final = final.concat([df, pd.DataFrame([df])], ignore_index=True)
```
